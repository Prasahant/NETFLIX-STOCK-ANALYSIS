# **NETFLIX STOCK ANALYSIS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv(r"F:\PYTHON FOR DA\z.CSV Files\Netflix.csv")
```

```python
data
```

```python
data.head()
```

```python
sns.set(rc={'figure.figsize': (10,5)})
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
data.head()
```

### Visualization of the volume of stock Traded in the year

```python
sns.lineplot(data, x= 'Date', y = 'Volume')
plt.title( 'Volume of Stock versus Time')
plt.show()
```

### Netflix Stock Price - High, Open, Close

```python
data.plot( y =['High','Close','Open'], title = 'Netflix Stock Price')
plt.show()
```

### Netflix Stock Price - Day, Month, Year Wise

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, figsize = (15, 11)) # it is used to create three small graphs
data.groupby(data.index.day).mean().plot(y='Volume', ax=ax1, xlabel = 'Day')
data.groupby(data.index.month).mean().plot(y='Volume', ax =ax2, xlabel = 'Month')
data.groupby(data.index.year).mean().plot(y='Volume', ax=ax3, xlabel="Year")
plt.show()
```

### Top-5 Dates with Highest Stock Price

```python
a = data.sort_values(by = 'High', ascending=False).head()
a["High"]
```

### Top-5 Dates with Lowest Stock Price

```python
b = data.sort_values(by = "Low", ascending=True).head()
b['Low']
```

### Visualize the highest and lowest trends in the graph 

```python
fig,axes = plt.subplots(nrows = 1, ncols=2, sharex = True, figsize =(15,4))
fig.suptitle("High and Low Values Stock per period of time", fontsize = 20)
sns.lineplot(ax = axes[0], y = data['High'], x = data.index, color = 'green')
sns.lineplot(ax=axes[1], y = data["Low"], x= data.index, color = 'Red')
plt.show()
```
